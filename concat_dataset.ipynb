```python
import pandas as pd

path_wiki = 'wikidata.csv'
path_dbpedia = 'dbpedia_query.csv'

# 2. Baca file CSV
df_wiki = pd.read_csv(path_wiki)
df_dbpedia = pd.read_csv(path_dbpedia)
```

```python
df_wiki = df_wiki.drop(columns=['event'])
```

```python
df_wiki_renamed = df_wiki.rename(columns={
    'eventLabel': 'nama_event',
    'caborLabel': 'cabang_olahraga',
    'lokasiLabel': 'lokasi',
    'tahun': 'tahun'
})

df_dbpedia_renamed = df_dbpedia.rename(columns={
    'eventName': 'nama_event',
    'cabangOlahraga': 'cabang_olahraga',
    'lokasiLengkap': 'lokasi',
    'tahun': 'tahun'
})

df_total = pd.concat([df_wiki_renamed, df_dbpedia_renamed], axis=0, ignore_index=True)

# 4. Cek Hasil
print(f"Baris Wiki: {len(df_wiki)}")
print(f"Baris DBpedia: {len(df_dbpedia)}")
print(f"Total Baris Setelah Concat: {len(df_total)}") 
```

```python
# 6. Simpan hasil final ke Drive
df_total.to_csv('df_concat.csv', index=False)
```
